```python
# Fill all null cells with the average 'Sentiment_Polarity' of all apps in the same Sentiment
df_copy['Sentiment_Polarity'] =df_copy.groupby('Sentiment')['Sentiment_Polarity'].transform(lambda x: x.fillna(x.mean()))
```

```python
# Fill all null cells with the average 'Sentiment_Subjectivity' of all apps in the same Sentiment
df_copy['Sentiment_Subjectivity'] =df_copy.groupby('Sentiment')['Sentiment_Subjectivity'].transform(lambda x: x.fillna(x.mean()))
```

```python
#Removing all the raows having Translated reviews as na
df_copy = df_copy.dropna(subset='Translated_Review')
```
