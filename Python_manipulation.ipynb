```python
import pandas as pd 

df =pd.read_csv("googleplaystore.csv")
df
```

```python
#sum of null values per column
df.isnull().sum()
```

# Task:1 Treatment Missing Values


```python
df = df.sort_values(by="Installs", ascending=False)
df.head()
```

```python
# Removing "Life Made WI-Fi Touchscreen Photo Frame" as it is having less installs

df.drop(df.index[0])
     
```

```python
# removing null values in Rating column
print(df.shape)
df.dropna(how='any',subset=["Rating"],axis=0,inplace=True)
```

```python
df.isnull().sum()
```

# Task: 2. Data Type correction

```python

#  remove $ sign in Price column
df["Price"] = df["Price"].apply(lambda x:x.replace("$","") if "$" in str(x)else x)

# Removing Row of 'Everyone' in Price Column
df.drop(labels=df[df['Price'] == 'Everyone'].index, inplace = True)

# Converting into Float
df["Price"] = df["Price"].apply(lambda x: float(x))
df
```

```python
# Removing "+" & "," in Install column
df['Installs'] = df['Installs'].apply(lambda x:x.replace("+","").replace(",","").strip() if "+" in str(x) else x)
df['Installs'] = df['Installs'].apply(lambda x:int(x))
df
```

```python
# Convert all other identified columns to numeric
df['Reviews'] = df['Reviews'].astype('int')
```

```python
df['Size'] = df['Size'].apply(lambda x: str(x).replace("M","").replace(",", "").replace("+", "").strip() if "M" in str(x) else x)

# Converting kb into Mb
df['Size'] = df['Size'].apply(lambda x:float(str(x).replace('k', '')) / 1024 if 'k' in str(x) else x)


# Removing Row of 'Varies with device' in Price Column
df.drop(labels=df[df['Size'] == 'Varies with device'].index, inplace = True)

# Converting into Float
df['Size'] = df["Size"].astype(float)
```

```python
# Drop any such record 

df.loc[df['Rating'] < 1]  & df.loc[df.Rating > 5]
df
```

```python
df.loc[df['Rating'] > df['Installs']]


Rating_Filter_data = df.loc[df['Rating'] > df['Installs']].index
df.drop(labels=Rating_Filter_data,inplace=True)
df
```

```python
df['Category'].value_counts()
```

```python
"""Comparison between Paid and Free apps. 
   - Find the most prevalent categories (based on number of apps)
   - Find the most installed categories (based on sum of installs)"""
num_of_apps_Free_apps = df[df['Type'] == 'Free']
num_of_apps_num_of_apps = num_of_apps_Free_apps['Category'].value_counts()
num_of_apps_num_of_apps
```

```python
num_of_apps_Paid_apps = df[df['Type'] == 'Paid']
num_of_apps_Paid_apps = num_of_apps_Paid_apps['Category'].value_counts()
num_of_apps_Paid_apps
```

```python
sum_of_installs_Paid_apps = df[df['Type'] == 'Paid']
sum_of_installs_Paid_apps  = sum_of_installs_Paid_apps.groupby('Category')[ 'Installs'].sum()
sum_of_installs_Paid_apps.sort_values(ascending=False)
```

```python
sum_of_installs_Free_apps = df[df['Type'] == 'Free']
sum_of_installs_Free_apps  = sum_of_installs_Free_apps.groupby('Category')[ 'Installs'].sum()
sum_of_installs_Free_apps.sort_values(ascending=False)
```

```python
# Sort the data by the 'Size' column in ascending order
df_sorted = df.sort_values('Size', ascending=True)

# Select the 50 smallest applications by size
df_smallest = df_sorted.head(50)

# Count the number of free and paid applications
num_free = df_smallest[df_smallest['Type'] == 'Free'].shape[0]
num_paid = df_smallest[df_smallest['Type'] == 'Paid'].shape[0]

# Print the results
print('Out of the 50 smallest applications by size:')
print(f'{num_free} are free')
print(f'{num_paid} are paid')
```
